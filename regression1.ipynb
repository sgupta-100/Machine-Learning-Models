# Working with the Regression Model

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X, y = fetch_california_housing(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()

lin.fit(X_train_scaled, y_train)
lin.score(X_test_scaled, y_test)
```

- Here .score is R^2 Score, not the accuracy.

- R^2 measures how close the data points are to the fitted regression line. 

- It is calculated by comparing your model's prediction errors to the errors you would get if you simply used the average (mean) of the dependent variable as a predictor.Range: 

- The score typically ranges from 0 to 1.

    - An R^2 of 1 means the model perfectly fits the data; all variance is explained.
    - An R^2 of 0 means the model explains none of the variability of the response data around its mean.
